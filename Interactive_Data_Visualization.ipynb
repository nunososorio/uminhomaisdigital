<h1>📊 Introdução às Visualizações Interativas</h1>
<p>
  As visualizações interativas são ferramentas que permitem ao utilizador explorar os dados de forma dinâmica, ajustando parâmetros, fazendo seleções 🎯 e filtrando informações em tempo real. Diferente dos gráficos estáticos, estas visualizações promovem uma experiência de análise mais profunda 🔎 e intuitiva, permitindo a descoberta de padrões e insights que podem não ser evidentes numa apresentação tradicional.
</p>

<h2>📖 Definições Relevantes</h2>
<ul>
  <li>
    <strong>📈 Visualização de Dados:</strong> O processo de representar informações e dados graficamente. O objetivo é facilitar a compreensão e a análise dos dados.
  </li>
  <li>
    <strong>🖱️ Visualização Interativa:</strong> Um tipo de visualização de dados que permite ao utilizador interagir com os gráficos, explorando os dados por meio de ações como zoom 🔍, filtros 🏷️ e seleções dinâmicas ✨. Essas interações ajudam a revelar informações ocultas e a compreender melhor a relação entre as variáveis.
  </li>
  <li>
    <strong>📊 Dashboards:</strong> Conjuntos de visualizações interativas dispostas de forma organizada que permitem a análise integrada de diversos aspectos dos dados.
  </li>
</ul>

<h2>🛠️ Conceitos Introdutórios</h2>
<p>
  Ao iniciar o estudo das visualizações interativas, é importante compreender:
</p>
<ul>
  <li><strong>🔄 Interatividade:</strong> A capacidade de um gráfico responder a ações do utilizador, como cliques e seleções, atualizando a informação mostrada.</li>
  <li><strong>🎨 Design e Usabilidade:</strong> A organização visual e a clareza dos elementos que tornam a visualização intuitiva e acessível.</li>
  <li><strong>📦 Ferramentas de Visualização:</strong> Plataformas e bibliotecas (como Plotly, Dash, etc.) que possibilitam a criação e customização de visualizações interativas com pouco código.</li>
  <li><strong>🔬 Exploração de Dados:</strong> Através das interações, o utilizador pode filtrar, agrupar e correlacionar variáveis, facilitando a descoberta de padrões e insights relevantes.</li>
</ul>


### 📊 Plotly: Visualização de Dados com Vários Componentes 🎨

**Plotly** é uma biblioteca poderosa para visualização de dados em Python, composta por diversos componentes que se complementam:

1.  **`plotly.graph_objects` (go):** 🛠️ Este é o núcleo da biblioteca. Permite um controlo detalhado sobre cada elemento do gráfico, ideal para personalização avançada e criação de visualizações complexas.

2.  **`plotly.express` (px):** 🚀 Uma interface de alto nível que simplifica a criação rápida de gráficos comuns com sintaxe concisa. Perfeito para exploração de dados e prototipagem rápida.

3.  **Plotly Dash:** 🌐 Um framework para construir aplicações web interativas e dashboards dinâmicos utilizando Plotly. Permite a criação de interfaces ricas e interativas para visualização de dados.

4.  **Plotly FigureFactory (ff):** 📈 Fornece funções para criar visualizações estatísticas complexas, como dendrogramas, heatmaps anotados e distribuições. Simplifica a criação de gráficos estatísticos avançados.

**Compartilhamento e Integração:**

* **Exportação para HTML (Offline):** 📦 Plotly permite exportar visualizações interativas para ficheiros HTML independentes. Estes ficheiros podem ser partilhados e visualizados em qualquer navegador, mesmo sem ligação à internet. Esta é a forma mais comum de partilhar os gráficos.

* **Integração com Jupyter:** 📓 Plotly integra-se perfeitamente com Jupyter Notebooks e JupyterLab, permitindo a criação e exibição de visualizações interativas diretamente nos notebooks. Esta integração é essencial para o fluxo de trabalho de análise de dados.

## Exemplo 1: Gráfico de Dispersão Interativo

Vamos criar um gráfico de dispersão que relaciona o PIB per capita (`gdpPercap`) com a expectativa de vida (`lifeExp`) para o ano de 2007, utilizando o dataset Gapminder. As cores representarão os continentes, o tamanho dos pontos representará a população e o hover mostrará o nome do país.

### Importação de bibliotecas e dados

```python
# Importar as bibliotecas necessárias
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Carregar um dataset de exemplo (Gapminder) usando Plotly
df = px.data.gapminder()
df
```

```python
# (opcional) usar o pandas describe para ter algumas estatísticas do df
print(df.describe())
```

```python
fig_scatter = px.scatter(
    df.query("year == 2007"),
    x='gdpPercap',
    y='lifeExp',
    color='continent',
    size='pop',
    hover_name='country',
    size_max=80,
    title='PIB per Capita vs Expectativa de Vida (2007)'
)
fig_scatter.show()
```

## Exemplo 2: Gráfico de Barras Animado

Neste exemplo, vamos visualizar a evolução da população por continente ao longo dos anos com um gráfico de barras animado.

```python
fig_bar = px.bar(
    df,
    x='country',
    y='pop',
    color='continent',
    animation_frame='year',
    title='Evolução da População por Continente'
)
fig_bar.show()
```

```python
# Criar uma lista com os países de interesse
ps = ['China', 'Portugal', 'Germany', 'Nigeria', 'United States', 'Australia']

# Filtrar o df para os países escolhidos
ps_df = df[df['country'].isin(ps)]

# Fazer o gráfico
fig_bar = px.bar(
    ps_df,
    x='country',
    y='pop',
    color='continent',
    animation_frame='year',
    title='Evolução da População por Continente'
)
fig_bar.show()
```

## Exemplo 3: Dashboard Simples com Subplots

Agora, vamos criar um dashboard simples que combine o gráfico de dispersão e um gráfico de barras (para um ano fixo) usando subplots do Plotly.

```python
from plotly.subplots import make_subplots
import plotly.express as px

ps = ['China', 'Portugal', 'Germany', 'Nigeria', 'United States', 'Australia']

# Criar um dashboard com 1 linha e 2 colunas
fig_dashboard = make_subplots(
    rows=1, cols=3,
    subplot_titles=("PIB per Capita vs Expectativa de Vida (2007)", "População por País (2007)", "Expectativa de Vida (2007)")
)

# Plot 1
fig_scatter = px.scatter(
    df.query("year == 2007"),
    x='gdpPercap',
    y='lifeExp',
    color='continent',
    size='pop',
    hover_name='country',
    size_max=80,
)

# Plot 2
fig_bar = px.bar(
    df.query("(year == 2007) & (country.isin(@ps))"),
    x='country',
    y='pop',
    color='continent',
)

# Plot 3
fig_bar2 = px.bar(
    df.query("(year == 2007) & (country.isin(@ps))"),
    x='country',
    y='lifeExp',
    color='continent',
)

# Add traces from fig_scatter to the dashboard
for trace in fig_scatter.data:
    fig_dashboard.add_trace(trace, row=1, col=1)

# Add traces from fig_bar to the dashboard
for trace in fig_bar.data:
    fig_dashboard.add_trace(trace, row=1, col=2)


# Add traces from fig_bar to the dashboard
for trace in fig_bar2.data:
    fig_dashboard.add_trace(trace, row=1, col=3)

# Update layout
fig_dashboard.update_layout(
    title_text='Dashboard Interativo Simples',
    showlegend=False
)

fig_dashboard.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# URL do conjunto de dados
dataset_url = 'https://raw.githubusercontent.com/plotly/datasets/master/2010_alcohol_consumption_by_country.csv'

# Ler os dados CSV diretamente do URL para um DataFrame pandas
df = pd.read_csv(dataset_url)

# Criar os dados do mapa coroplético
data = [{
    'type': 'choropleth',
    'locationmode': 'country names',
    'locations': df['location'],
    'z': df['alcohol'],
    'text': df['location'],
    'autocolorscale': True
}]

# Definir o layout do mapa
layout = {
    'title': {'text': 'Consumo de álcool<br>entre adultos (idade 15+) em 2010'},
    'geo': {
        'projection': {
            'type': 'robinson'
        }
    }
}

# Criar e exibir a figura
fig = go.Figure(data=data, layout=layout)
fig.show()

# Notas:
# 1. Este código cria um mapa coroplético que mostra o consumo de álcool puro por país em 2010.
# 2. Os dados são carregados diretamente de um URL usando pandas.
# 3. Plotly graph_objects é usado para criar o mapa com controlo detalhado.
# 4. O mapa é exibido interativamente no navegador.
```

```python
import pandas as pd
import plotly.express as px

# URL do conjunto de dados
dataset_url = 'https://raw.githubusercontent.com/plotly/datasets/master/2010_alcohol_consumption_by_country.csv'

# Ler os dados CSV diretamente do URL para um DataFrame pandas
df = pd.read_csv(dataset_url)

# Criar o mapa coroplético com Plotly Express
fig = px.choropleth(df,
                    locations='location',
                    locationmode='country names',
                    color='alcohol',
                    title='Consumo de álcool entre adultos (idade 15+) em 2010',
                    projection='robinson')

# Exibir a figura
fig.show()

# Notas:
# 1. Este código cria um mapa coroplético usando Plotly Express, que simplifica o processo.
# 2. As colunas 'location' e 'alcohol' do DataFrame são usadas diretamente.
# 3. O parâmetro 'projection' define a projeção do mapa para 'robinson'.
# 4. Plotly Express lida com a maior parte da configuração automaticamente.
# 5. O resultado é um mapa interativo exibido no navegador.
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

# URL do conjunto de dados
dataset_url = 'https://raw.githubusercontent.com/plotly/datasets/master/2010_alcohol_consumption_by_country.csv'

# Ler os dados CSV diretamente do URL para um DataFrame pandas
df = pd.read_csv(dataset_url)

# Criar o mapa coroplético com Plotly Express
fig = px.choropleth(df,
                    locations='location',
                    locationmode='country names',
                    color='alcohol',
                    title='Consumo de álcool puro entre adultos (idade 15+) em 2010',
                    projection='robinson',
                    hover_name='location',  # Adiciona o nome do país no hover
                    hover_data={'alcohol': ':.2f'}) #formats the hover data for alcohol.

# Exportar o mapa para um ficheiro HTML
pio.write_html(fig, 'alcohol_consumption_map_hover.html')

print("Mapa exportado para alcohol_consumption_map_hover.html")
```

<div>
  <h3>🗺️ Mapa Coroplético (Choropleth Map) Explicado</h3>

  <p>
    Um <b>mapa coroplético</b> utiliza cores para representar dados geográficos. A intensidade da cor indica a variação de uma medida entre diferentes regiões.
  </p>

  <p>
    <b>Projeção de Robinson:</b> Uma forma de "achatar" a Terra para criar um mapa plano, equilibrando distorção de áreas e formas.
  </p>

  <p>
    <b>Parâmetros de dados para Plotly (Mapa Coroplético):</b>
  </p>

  <ul>
    <li>
      <b><code>locations</code>:</b> 📍 Obrigatório. Especifica as áreas geográficas (nomes de países, códigos de estados, etc.).
    </li>
    <li>
      <b><code>locationmode</code>:</b> 🌍 Obrigatório. Define como as localizações são identificadas (ex: 'country names', 'ISO-3').
    </li>
    <li>
      <b><code>color</code>:</b> 🎨 Obrigatório. Determina a coluna de dados que define a cor de cada área (a variável que se quer mostrar no mapa).
    </li>
    <li>
      <b>Outros parâmetros:</b> (opcionais) como <code>projection</code> (tipo de projeção do mapa), <code>title</code> (título do mapa), <code>hover_name</code> (texto exibido ao passar o rato sobre as áreas), entre outros.
    </li>
  </ul>
</div>

```python

import pandas as pd
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Load data
dataset_url = 'https://raw.githubusercontent.com/plotly/datasets/master/2010_alcohol_consumption_by_country.csv'
df = pd.read_csv(dataset_url)

# Create the initial figure
fig = px.choropleth(df,
                    locations='location',
                    locationmode='country names',
                    color='alcohol',
                    title='Consumo de álcool puro entre adultos (idade 15+) em 2010',
                    projection='robinson',
                    hover_name='location',
                    hover_data={'alcohol': ':.2f'})

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the app layout
app.layout = html.Div([
    html.H1("Alcohol Consumption Map Dashboard"),
    dcc.Graph(id='alcohol-map', figure=fig),
    html.Div([
        html.Label("Change Projection:"),
        dcc.Dropdown(
            id='projection-dropdown',
            options=[
                {'label': 'Robinson', 'value': 'robinson'},
                {'label': 'Orthographic', 'value': 'orthographic'},
                {'label': 'Natural Earth', 'value': 'natural earth'}
            ],
            value='robinson'
        )
    ])
])

# Define the callback to update the figure
@app.callback(
    Output('alcohol-map', 'figure'),
    [Input('projection-dropdown', 'value')]
)
def update_map(selected_projection):
    updated_fig = px.choropleth(df,
                                locations='location',
                                locationmode='country names',
                                color='alcohol',
                                title='Consumo de álcool puro entre adultos (idade 15+) em 2010',
                                projection=selected_projection,
                                hover_name='location',
                                hover_data={'alcohol': ':.2f'})
    return updated_fig

# Run the app
if __name__ == '__main__':
    app.run_server(mode='inline', port=8050)
```
